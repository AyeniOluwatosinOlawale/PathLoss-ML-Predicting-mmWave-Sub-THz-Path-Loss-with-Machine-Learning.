```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import TransformedTargetRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import learning_curve
import time
from sklearn.preprocessing import MinMaxScaler
from sklearn.dummy import DummyRegressor
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform
from sklearn.tree import DecisionTreeRegressor
import os
from keras.models import Sequential
from keras.layers import Dense
```

```python
X = data.loc[:,data.columns != 'Path Loss (dB)']
y = data.loc[:,'Path Loss (dB)']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=18)
```

# Modelling and Evaluation of the peformance of the models

```python
#Creating a baseline Model

dummy_regressor = DummyRegressor(strategy='mean')  
dummy_regressor.fit(X_train, y_train)

# Making predictions using the dummy regressor
y_pred = dummy_regressor.predict(X_test)
```

```python
#Dummy regressor Evaluation
mse_5 = mean_squared_error(y_test, y_pred)
rmse_5 = np.sqrt(mse_5)
r2_5 = r2_score(y_test, y_pred)
mae_5 = mean_absolute_error(y_test, y_pred)
print("dummy_regressor MAE:", mae_5)
print("dummy_regressor R-squared:", r2_5)
print("dummy_regressor RMSE:", rmse_5)
print("dummy_regressor MSE:", mse_5)
```

```python

def plot_metrics(mse_values, rmse_values, r2_values, mae_values, model_names, output_dir):
    # Creating DataFrames for each metric
    MSE = pd.DataFrame({model_names[i]: mse_values[i] for i in range(len(model_names))}, index=['MSE'])
    RMSE = pd.DataFrame({model_names[i]: rmse_values[i] for i in range(len(model_names))}, index=['RMSE'])
    R2 = pd.DataFrame({model_names[i]: r2_values[i] for i in range(len(model_names))}, index=['R2'])
    MAE = pd.DataFrame({model_names[i]: mae_values[i] for i in range(len(model_names))}, index=['MAE'])

    # Plotting the bar charts for each metric
    fig, axes = plt.subplots(2, 2, figsize=(15, 15))
    sns.barplot(data=MSE, ax=axes[0, 0])
    axes[0, 0].set(xlabel='Model', ylabel='MSE')
    axes[0, 0].set_title('Mean Square Error (MSE) by Models for UmI Scenario')

    sns.barplot(data=RMSE, ax=axes[0, 1])
    axes[0, 1].set(xlabel='Model', ylabel='RMSE')
    axes[0, 1].set_title('Root Mean Square Error (RMSE) by Models for Umi Scenario')

    sns.barplot(data=R2, ax=axes[1, 0])
    axes[1, 0].set(xlabel='Model', ylabel='R2')
    axes[1, 0].set_title('R2 (R2) by Models for Umi Scenario')

    sns.barplot(data=MAE, ax=axes[1, 1])
    axes[1, 1].set(xlabel='Model', ylabel='MAE')
    axes[1, 1].set_title('Mean Square Error (MAE) by Models for Umi Scenario')

    plt.tight_layout()
    plt.show()

    # Creating the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Saving the figures as JPEG image files
    fig.savefig(f'{output_dir}/metrics_plot.jpg')
    plt.close()

# MSE, RMSE, R2, and MAE values for each models
mse_values = [mse_1, mse_2, mse_3, mse_4, mse_5]
rmse_values = [rmse_1, rmse_2, rmse_3, rmse_4, rmse_5]
r2_values = [r2_1, r2_2, r2_3, r2_4, r2_5]
mae_values = [mae_1, mae_2, mae_3, mae_4, mae_5]
model_names = ['Random Forest', 'Gradient Boost', 'Ada Boost', 'Xgb Boost', 'Dummy Regressor']
output_dir = 'New folder1' 

plot_metrics(mse_values, rmse_values, r2_values, mae_values, model_names, output_dir)
```

```python

def plot_learning_curve(model, X_train, y_train, cv=5, model_name=None, ax=None):
    # Defining the training set sizes for the learning curve
    train_sizes, train_scores, test_scores = learning_curve(
        model, X_train, y_train, cv=cv, train_sizes=np.linspace(0.1, 1.0, 10), scoring='neg_mean_squared_error'
    )

    # Calculating the mean and standard deviation of training and test scores
    train_mean = -np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = -np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    # Plotting the learning curve on the provided axis
    ax.plot(train_sizes, np.sqrt(train_mean), label='Training RMSE', color='blue')
    ax.plot(train_sizes, np.sqrt(test_mean), label='Validation RMSE', color='red')

    # Adding standard deviation regions
    ax.fill_between(train_sizes, np.sqrt(train_mean - train_std), np.sqrt(train_mean + train_std), color='blue', alpha=0.1)
    ax.fill_between(train_sizes, np.sqrt(test_mean - test_std), np.sqrt(test_mean + test_std), color='red', alpha=0.1)

    # Adding model name as annotation
    last_train_mean = np.sqrt(train_mean[-1])
    last_test_mean = np.sqrt(test_mean[-1])
    ax.annotate(model_name, xy=(train_sizes[-1], last_train_mean), xytext=(-20, -10),
                textcoords='offset points', ha='right', va='center', color='blue', weight='bold')
    ax.annotate(model_name, xy=(train_sizes[-1], last_test_mean), xytext=(-20, 10),
                textcoords='offset points', ha='right', va='center', color='red', weight='bold')

def save_learning_curve_plot(models, X_train, y_train, output_dir):
    # Creating a single figure to combine all learning curve plots
    plt.figure(figsize=(12, 8))

    # Creating a single subplot for the learning curve
    ax = plt.subplot(1, 1, 1)
    for model, model_name in models:
        plot_learning_curve(model, X_train, y_train, model_name=model_name, ax=ax)

    # Adding labels, title, and legend to the figure
    plt.xlabel('Training Set Size')
    plt.ylabel('Root Mean Squared Error (RMSE)')
    plt.title('Learning Curves for Different Models in Umi Scenario')
    plt.legend(loc='best')

    # Saving the figure as JPEG
    plt.savefig(f'{output_dir}/learning_curves.jpg', format='jpeg')

    # Displaying the plot
    plt.tight_layout()
    plt.show()



#displaying the learning curve plot
output_dir = 'New folder' 
models = [(rf_best, 'Random Forest'),
          (xgb_best, 'XGBoost'),
          (adaboost_best, 'AdaBoost'),
          (gb_best1, 'Gradient Boost')]
save_learning_curve_plot(models, X_train, y_train, output_dir)
```
